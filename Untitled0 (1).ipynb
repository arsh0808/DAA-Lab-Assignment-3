```python

!pip install memory-profiler matplotlib > /dev/null

import time
from memory_profiler import memory_usage
import matplotlib.pyplot as plt
import heapq
from collections import defaultdict, deque


def suggest_friends(graph, user):
    visited = set()
    queue = deque([user])
    visited.add(user)
    direct_friends = set(graph[user])
    suggestions = defaultdict(int)

    while queue:
        current = queue.popleft()
        for friend in graph[current]:
            if friend not in visited:
                visited.add(friend)
                queue.append(friend)
            if friend not in direct_friends and friend != user:
                suggestions[friend] += 1

    sorted_suggestions = sorted(suggestions.items(), key=lambda x: -x[1])
    return [u for u, _ in sorted_suggestions]


def run_social_network():
    graph = {
        'A': ['B', 'C'],
        'B': ['A', 'D'],
        'C': ['A', 'D', 'E'],
        'D': ['B', 'C'],
        'E': ['C']
    }
    user = 'A'
    print("üë• Suggested friends for", user, ":", suggest_friends(graph, user))



def bellman_ford(edges, vertices, source):
    distance = {v: float('inf') for v in vertices}
    distance[source] = 0

    for _ in range(len(vertices) - 1):
        for u, v, w in edges:
            if distance[u] + w < distance[v]:
                distance[v] = distance[u] + w

    for u, v, w in edges:
        if distance[u] + w < distance[v]:
            print("‚ö†Ô∏è Graph contains negative weight cycle")
            return None

    return distance


def run_bellman_ford():
    vertices = ['A', 'B', 'C', 'D']
    edges = [
        ('A', 'B', 4),
        ('A', 'C', 2),
        ('C', 'B', -1),
        ('B', 'D', 2),
        ('C', 'D', 3)
    ]
    result = bellman_ford(edges, vertices, 'A')
    print("üó∫Ô∏è Shortest distances from A:", result)



def dijkstra(graph, source):
    distances = {node: float('inf') for node in graph}
    distances[source] = 0
    pq = [(0, source)]

    while pq:
        current_dist, current_node = heapq.heappop(pq)
        if current_dist > distances[current_node]:
            continue
        for neighbor, weight in graph[current_node]:
            distance = current_dist + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))
    return distances


def run_dijkstra():
    graph = {
        'A': [('B', 1), ('C', 4)],
        'B': [('C', 2), ('D', 5)],
        'C': [('D', 1)],
        'D': []
    }
    print("üöë Fastest routes from A:", dijkstra(graph, 'A'))



def prims_mst(graph, start):
    visited = set()
    pq = [(0, start)]
    total_cost = 0

    while pq:
        weight, node = heapq.heappop(pq)
        if node not in visited:
            visited.add(node)
            total_cost += weight
            for neighbor, w in graph[node]:
                if neighbor not in visited:
                    heapq.heappush(pq, (w, neighbor))
    return total_cost


def run_prims():
    graph = {
        'A': [('B', 2), ('C', 3)],
        'B': [('A', 2), ('C', 1), ('D', 4)],
        'C': [('A', 3), ('B', 1), ('D', 5)],
        'D': [('B', 4), ('C', 5)]
    }
    print("üè¢ Total minimum cable cost:", prims_mst(graph, 'A'))



def profile_algorithm(func, *args):
    start_time = time.time()
    mem_usage = memory_usage((func, args), max_iterations=1)
    end_time = time.time()
    return (end_time - start_time, max(mem_usage))


def run_profiling():
    algorithms = [
        ("Social Network", run_social_network),
        ("Bellman-Ford", run_bellman_ford),
        ("Dijkstra", run_dijkstra),
        ("Prim‚Äôs MST", run_prims)
    ]

    times, mems, names = [], [], []

    for name, func in algorithms:
        print(f"\n===== Profiling {name} =====")
        t, m = profile_algorithm(func)
        print(f"‚è± Time: {t:.6f}s | üíæ Memory: {m:.2f} MB")
        times.append(t)
        mems.append(m)
        names.append(name)

    plt.figure(figsize=(8, 5))
    plt.bar(names, times, color='skyblue')
    plt.title("Execution Time of Graph Algorithms")
    plt.ylabel("Time (seconds)")
    plt.show()

    plt.figure(figsize=(8, 5))
    plt.bar(names, mems, color='lightcoral')
    plt.title("Memory Usage of Graph Algorithms")
    plt.ylabel("Memory (MB)")
    plt.show()



def summary_table():
    print("\nüìÑ FINAL SUMMARY")
    print("---------------------------------------------------------------")
    print(f"{'Problem':<25}{'Algorithm':<15}{'Time Complexity':<15}{'Domain'}")
    print("---------------------------------------------------------------")
    print(f"{'Social Network Suggestion':<25}{'BFS / DFS':<15}{'O(V + E)':<15}{'Social Media'}")
    print(f"{'Route Finding':<25}{'Bellman-Ford':<15}{'O(VE)':<15}{'Navigation'}")
    print(f"{'Emergency Response':<25}{'Dijkstra':<15}{'O(E log V)':<15}{'Disaster Mgmt'}")
    print(f"{'Cable Installation':<25}{'Prim‚Äôs/Kruskal':<15}{'O(E log V)':<15}{'Infrastructure'}")
    print("---------------------------------------------------------------")


def main():
    print("=== GRAPH ALGORITHMS PROJECT (COLAB VERSION) ===")
    run_social_network()
    run_bellman_ford()
    run_dijkstra()
    run_prims()
    run_profiling()
    summary_table()

main()
```
